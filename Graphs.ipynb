```python
import json
import matplotlib.pyplot as plt

# Load the JSON data from the file
data = []
with open('output.jsonl', 'r') as file:
    for line in file:
        data.append(json.loads(line))

# Extract the toxicity scores
overall_toxicity_scores = [entry["models_out"][0]["com_toxicity"] for entry in data]
paraphrase_toxicity_scores = [entry["paraphrase"]["com_toxicity"] for entry in data]

# Create the plot
plt.figure(figsize=(10, 6))
plt.plot(overall_toxicity_scores, label='Toxicity (MIL-Decoding)', marker='o')
plt.plot(paraphrase_toxicity_scores, label='Toxicity (Paradetox)', marker='x')
plt.xlabel('Sentence Index')
plt.ylabel('Toxicity Score')
plt.title('Comparison of Toxicity Scores')
plt.legend()
plt.grid(True)
plt.show()
```

```python
import json
import matplotlib.pyplot as plt

# Load the JSON data from the file
data = []
with open('output.jsonl', 'r') as file:
    for line in file:
        data.append(json.loads(line))

# Extract the GPT perplexity scores
mil_decoding_gpt_ppl_scores = [entry["models_out"][0]["calc_gpt_ppl"] for entry in data]
paradetox_gpt_ppl_scores = [entry["paraphrase"]["calc_gpt_ppl"] for entry in data]

# Create the plot
plt.figure(figsize=(10, 6))
plt.plot(mil_decoding_gpt_ppl_scores, label='GPT Perplexity (MIL-Decoding)', marker='o')
plt.plot(paradetox_gpt_ppl_scores, label='GPT Perplexity (Paradetox)', marker='x')
plt.xlabel('Sentence Index')
plt.ylabel('GPT Perplexity Score')
plt.title('Comparison of GPT Perplexity Scores')
plt.legend()
plt.grid(True)
plt.show()
```

```python
import numpy as np

# Extract the scores for models_out
models_out_diversity = [entry["models_out"][0]["diversity"] for entry in data]
models_out_gpt_ppl = [entry["models_out"][0]["calc_gpt_ppl"] for entry in data]
models_out_com_toxicity = [entry["models_out"][0]["com_toxicity"] for entry in data]

# Extract the scores for paraphrase
paraphrase_diversity = [entry["paraphrase"]["diversity"] for entry in data]
paraphrase_gpt_ppl = [entry["paraphrase"]["calc_gpt_ppl"] for entry in data]
paraphrase_com_toxicity = [entry["paraphrase"]["com_toxicity"] for entry in data]

# Calculate the means
mean_models_out_diversity = np.mean(models_out_diversity)
mean_models_out_gpt_ppl = np.mean(models_out_gpt_ppl)
mean_models_out_com_toxicity = np.mean(models_out_com_toxicity)

mean_paraphrase_diversity = np.mean(paraphrase_diversity)
mean_paraphrase_gpt_ppl = np.mean(paraphrase_gpt_ppl)
mean_paraphrase_com_toxicity = np.mean(paraphrase_com_toxicity)

# Print the results
print("Mean Diversity (models_out):", mean_models_out_diversity)
print("Mean GPT Perplexity (models_out):", mean_models_out_gpt_ppl)
print("Mean Com Toxicity (models_out):", mean_models_out_com_toxicity)

print("Mean Diversity (paraphrase):", mean_paraphrase_diversity)
print("Mean GPT Perplexity (paraphrase):", mean_paraphrase_gpt_ppl)
print("Mean Com Toxicity (paraphrase):", mean_paraphrase_com_toxicity)
```
